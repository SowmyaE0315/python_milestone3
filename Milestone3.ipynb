```python
filename='covid_usa.xlsx'
df=pd.read_excel(filename)
```

```python
df.head()
```

```python
states=df.loc[:,'state'].values
cases=df.loc[:,'total_cases'].values
```

```python
df['total_cases'].hist()
plt.yticks(rotation=90)
```

```python
plt.bar(states,cases)
```

```python
plt.figure(figsize=(40,10))
plt.bar(states,cases,width=0.9,align='center',color='cyan',edgecolor='red')
i=2.0
j=2000
for i in range(len(states)):
    plt.annotate(cases[i], (-0.1 + i,cases[i] +j))
    
    
plt.title('plot for cases by state')
plt.xlabel('states')
plt.ylabel('cases')
plt.savefig('images/CasesByState',dpi=500)
plt.show()    
```

```python
plt.figure(figsize=(40,10))
states=df.loc[:,'state'].values
recovered=df.loc[:,'total_recovered'].values
  
plt.bar(states,recovered,width=0.9,align='center',color='cyan',edgecolor='red')
i=2.0
j=2000
for i in range(len(states)):
    plt.annotate(recovered[i], (-0.1 + i,recovered[i] +j))
    
    
plt.title('plot for cases by state')
plt.xlabel('states')
plt.ylabel('recovered')
plt.savefig('images/RecoveryByState',dpi=500)
plt.show()     
```

```python
plt.figure(figsize=(12,7))
countries=df1.loc[:,'Country'].values
topcases=df1.loc[:,'Total_Cases'].values
plt.bar(countries,topcases,width=0.5,align='center',color='cyan',edgecolor='red')
i=0.5
j=500
for i in range(len(countries)):
    plt.annotate(topcases[i], (-0.1 + i,topcases[i] +j))
    
    
plt.title('plot for cases by top5 countries')
plt.xlabel('Countries')
plt.ylabel('topcases')
plt.savefig('images/Top5Countries',dpi=500)
plt.show()     
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

data='music.csv'
df=pd.read_csv(data)
x=df.drop(columns=['genre'])
y=df['genre']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)

model1=DecisionTreeClassifier()
model1.fit(x_train,y_train)
prediction=model1.predict(x_test)
score=accuracy_score(y_test,prediction)
score
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score 
import joblib

data='music.csv'
df=pd.read_csv(data)
x=df.drop(columns=['genre'])
y=df['genre']

model1=DecisionTreeClassifier()
model1.fit(x,y)
joblib.dump(model1,'music-recommender.joblib')
```

```python
model1=joblib.load('music-recommender.joblib')
prediction=model1.predict([[21,1]])
prediction
```
